```python
# Design model
#  Construct loss and optimizer
# Training Loop
#         - forward pass :compute prediction and loss
#         - backward pass :gradients
#         - update weights 
import torch
import torch.nn as nn
import numpy
from sklearn import datasets 
import matplotlib.pyplot as plt
```

```python
# Preparing our dataset 
X_numpy, y_numpy=datasets.make_regression(n_samples=100,n_features=1,noise=20,random_state=1)
X=torch.from_numpy(X_numpy.astype(numpy.float32))
y=torch.from_numpy(y_numpy.astype(numpy.float32)).view(-1,1)
n_samples,n_features=X.shape

# Defining our model
class LinearModel(nn.Module):
    def __init__(self,input_size,output_size):
        super().__init__()
        self.linear=nn.Linear(input_size,output_size)
    
    def forward(self,x):
        return self.linear(x)
    
input_size=n_features
output_size=1
model=LinearModel(input_size,output_size)
lr=0.01
loss=nn.MSELoss()
optimizer=torch.optim.SGD(model.parameters(),lr=lr)
n_iterations=500
losses=[]
for epoch in range(n_iterations):
    model.train()
    y_pred=model(X)
    l=loss(y_pred,y)
    losses.append(l.item())
    #backward pass
    l.backward()
    # update
    optimizer.step()#optimizer.step() applies these gradients to the parameters, updating the model's weights.
    # clear the old gradients
    optimizer.zero_grad()
    if (epoch%2==0):
        print(f'epoch{epoch+1},loss={l:.8f}')
predicted=model(X).detach().numpy()
plt.plot(X_numpy,y_numpy,"ro")
plt.plot(X_numpy,predicted,"b")
plt.show()
    

```
